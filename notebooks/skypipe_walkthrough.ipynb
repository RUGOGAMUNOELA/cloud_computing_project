# SkyPipe Walkthrough

Run from repo root with `PYTHONPATH=src` or `cd src` before executing cells.

Demonstrates **adaptive analytics** on arbitrary CSV without hardcoded columns.

```python
import os, sys, json
ROOT = os.path.abspath(os.path.join(os.getcwd(), "..", "src"))
if ROOT not in sys.path:
    sys.path.insert(0, ROOT)

from spark_pipeline import build_spark_session, load_dataframe, adaptive_analytics
from schema_detector import build_schema_profile

sample = os.path.join(ROOT, "data", "sample_sales.csv")
spark = build_spark_session("Notebook-Demo")
df = load_dataframe(spark, sample, "csv")
profile = build_schema_profile(df)
print("Profile:", profile)
summary = adaptive_analytics(df, profile=profile)
spark.stop()
print(json.dumps(summary, indent=2, default=str))
```
